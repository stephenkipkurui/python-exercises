# #   --> How many different combinations are there of 2 letters from "abcd"?


```python
from itertools import combinations

letter = 'abcd'

letter_2_combinations = combinations(letter, 2)

ltr = [' | '.join(i) for i in letter_2_combinations]

print(ltr)
```
